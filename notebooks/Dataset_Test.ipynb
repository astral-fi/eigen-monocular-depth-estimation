```python
import sys
import os

PROJECT_ROOT = os.path.abspath("..")
sys.path.append(PROJECT_ROOT)

print("Added to path:", PROJECT_ROOT)
from models.dataset import Dataset
import matplotlib.pyplot as plt

```

```python
dataset = Dataset(
    root_dir="../data/",
    img_size=(224, 224)
)

print("Dataset size:", len(dataset))
print(dataset.img_dir)
```

```python
sample = dataset[0]

img = sample["image"]
depth = sample["depth"]
mask = sample["mask"]

print(img.shape, img.dtype)
print(depth.shape, depth.dtype)
print(mask.shape, mask.dtype)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.imshow(img.permute(1,2,0))
plt.title("RGB (tensor)")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(depth.squeeze(), cmap="plasma")
plt.title("Depth (tensor, meters)")
plt.colorbar()
plt.axis("off")

plt.show()
```

```python
from torch.utils.data import DataLoader

loader = DataLoader(
    dataset,
    batch_size=4,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

batch = next(iter(loader))

print(batch["image"].shape)
print(batch["depth"].shape)
```
