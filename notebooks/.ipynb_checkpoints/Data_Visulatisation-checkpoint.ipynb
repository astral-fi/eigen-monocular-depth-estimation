```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2

```

```python
IMG_DIR = "../data/images_paired"
DEPTH_DIR = "../data/depths_paired"

files = sorted(os.listdir(IMG_DIR))
print("Number of paired samples:", len(files))
```

```python
idx = 0

fname = files[idx]

img = np.array(
    Image.open(os.path.join(IMG_DIR, fname)).convert("RGB")
)

depth_raw = np.array(
    Image.open(os.path.join(DEPTH_DIR, fname))
)

print("RGB shape:", img.shape, img.dtype)
print("Depth shape:", depth_raw.shape, depth_raw.dtype)
print("Depth min/max:", depth_raw.min(), depth_raw.max())
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("RGB Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(depth_raw, cmap="plasma")
plt.title("Raw Depth (uint16)")
plt.colorbar()
plt.axis("off")

plt.show()
```

```python
depth_m = depth_raw.astype(np.float32) / 256.0
print("Depth (meters) min/max:", depth_m.min(), depth_m.max())

valid_mask = depth_m > 0
depth_m[~valid_mask] = 0

plt.imshow(depth_m, cmap="plasma")
plt.title("Depth in meters (masked)")
plt.colorbar()
plt.axis("off")
plt.show()
```

```python
TARGET_SIZE = (224, 224)

img_resized = cv2.resize(img, TARGET_SIZE)
depth_resized = cv2.resize(
    depth_m, TARGET_SIZE, interpolation=cv2.INTER_NEAREST
)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.imshow(img_resized)
plt.title("RGB (224224)")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(depth_resized, cmap="plasma")
plt.title("Depth (224224, meters)")
plt.colorbar()
plt.axis("off")

plt.show()
```
