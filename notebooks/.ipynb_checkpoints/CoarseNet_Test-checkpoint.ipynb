```python
import torch
import matplotlib.pyplot as plt
from models.dataset import Dataset
from models.coarse_net import CoarseDepthNet

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

dataset = Dataset("../data/")
model = CoarseDepthNet().to(DEVICE)

model.load_state_dict(torch.load("../checkpoints/coarse_epoch_10.pth"))
model.eval()

sample = dataset[0]
img = sample["image"].unsqueeze(0).to(DEVICE)

with torch.no_grad():
    pred = model(img)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(sample["image"].permute(1,2,0))
plt.title("RGB")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(sample["depth"].squeeze(), cmap="plasma")
plt.title("GT Depth")
plt.colorbar()

plt.subplot(1,3,3)
plt.imshow(pred.squeeze().cpu(), cmap="plasma")
plt.title("Predicted Depth (Coarse)")
plt.colorbar()

plt.show()
```
